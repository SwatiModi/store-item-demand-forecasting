```python
df['Timestamp'] = pd.to_datetime(df.index,format='%Y-%m-%d') 
df.index = df.Timestamp
```

```python
Train = df.loc['2013-01-01':'2016-12-31'] 
valid = df.loc['2016-12-31':'2017-12-31']
```

```python
Train.shape
```

```python
valid.shape
```

```python
Train.sales.plot(figsize=(15,8), title= 'Daily Sales', fontsize=14, label='train') 
valid.sales.plot(figsize=(15,8), title= 'Daily Sales', fontsize=14, label='valid') 
plt.xlabel("Datetime") 
plt.ylabel("Sales") 
plt.legend(loc='best') 
plt.show()
```

Naive Approach

```python
dd = np.asarray(Train.sales)
```

```python
y_hat = valid.copy()
```

```python
y_hat['naive'] = dd[len(dd)-1]
```

```python
plt.figure(figsize=(12,8)) 
plt.plot(Train.index, Train['sales'], label='Train')
plt.plot(valid.index, valid['sales'], label='valid')
plt.plot(y_hat.index, y_hat['naive'], label='Naive forecast values')
plt.legend(loc='best') 
plt.title("Naive Forecast") 
plt.show()
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
```

```python
rmse = sqrt(mean_squared_error(valid.sales,y_hat.naive))
```

```python
rmse
```

##### Moving average

```python
Train['sales'].rolling(600).mean()[-1]
```

```python
y_hat = valid.copy()
```

```python
y_hat['Moving Average'] = Train['sales'].rolling(600).mean()[-1]
```

```python
plt.figure(figsize=(12,8)) 
plt.plot(Train.index, Train['sales'], label='Train')
plt.plot(valid.index, valid['sales'], label='valid')
plt.plot(y_hat.index, y_hat['Moving Average'], label='Moving average forecast values')
plt.legend(loc='best') 
plt.title("Moving average Forecast") 
plt.show()
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
```

```python
rmse = sqrt(mean_squared_error(valid.sales,y_hat['Moving Average']))
```

```python
rmse
```
