```python
data = df.copy()
```

```python
data.index = pd.to_datetime(data.index, format='%Y-%m-%d')
```

```python
data['day'] = data.index.day
data['month'] = data.index.month
data['year'] = data.index.year
data['dayofweek'] = data.index.dayofweek
```

```python
data
```

```python
year_2013 = data[data['year']==2013]
year_2014 = data[data['year']==2014]
year_2015 = data[data['year']==2015]
year_2016 = data[data['year']==2016]
year_2017 = data[data['year']==2017]
```

```python
data.groupby('dayofweek')['sales'].mean().plot.bar()
```

We can see here that the sales have an increasing trend in sales weekwise

```python
def weekend(row):
    if row == 5 or row == 6:
        return 1
    else:
        return 0
```

```python
temp2 = data['dayofweek'].apply(weekend) 
```

```python
data['weekend'] = temp2
```

```python
data.groupby('weekend')['sales'].mean().plot.bar()
```

```python
daily = data['day'].resample('D').mean()
monthly = data['month'].resample('M').mean()
```

```python
fig, axs = plt.subplots(2,1) 
daily.plot(figsize=(15,8), title= 'Daily', fontsize=12, ax=axs[0]) 
monthly.plot(figsize=(15,8), title= 'Monthly', fontsize=12, ax=axs[1]) 
```
