```python
f, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(5, figsize = (12, 13))
c = '#386B7F'

# store types
sales_1.plot(color = c, ax = ax1)
sales_2.plot(color = c, ax = ax2)
sales_3.plot(color = c, ax = ax3)
sales_4.plot(color = c, ax = ax4)
sales_5.plot(color = c, ax = ax5)
```

```python
f, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(5, figsize = (12, 13))
import statsmodels.api as sm

decomposition_a = sm.tsa.seasonal_decompose(sales_1, model = 'additive', freq = 365)
decomposition_a.trend.plot(color = c, ax = ax1)

decomposition_a = sm.tsa.seasonal_decompose(sales_2, model = 'additive', freq = 365)
decomposition_a.trend.plot(color = c, ax = ax2)

decomposition_a = sm.tsa.seasonal_decompose(sales_3, model = 'additive', freq = 365)
decomposition_a.trend.plot(color = c, ax = ax3)

decomposition_a = sm.tsa.seasonal_decompose(sales_4, model = 'additive', freq = 365)
decomposition_a.trend.plot(color = c, ax = ax4)

decomposition_a = sm.tsa.seasonal_decompose(sales_5, model = 'additive', freq = 365)
decomposition_a.trend.plot(color = c, ax = ax5)
```
